## Delta accuracy

```python
def remove_outliers(df, columns, quantile):
    for column in columns:
        df.loc[df[column] > df[column].quantile(quantile), column] = np.nan
    return df

delta_hagan = pd.read_pickle(data_path + '/simulation_delta_hagan.pkl')
delta_ANN = pd.read_pickle(data_path + '/simulation_delta_ANN.pkl')

simulation_hagan = pd.read_pickle(data_path + '/simulation_dict_hagan.pkl')
simulation_ann = pd.read_pickle(data_path + '/simulation_dict_ann.pkl')
simulation_antonov = pd.read_pickle(data_path + '/simulation_dict_antonov.pkl')


delta_hagan = simulation_hagan['delta_error']
delta_ANN = simulation_ann['delta_error']
delta_antonov = simulation_antonov['delta_error']


q = .995
delta_hagan = remove_outliers(delta_hagan, delta_hagan.columns, q)
delta_ANN = remove_outliers(delta_ANN, delta_ANN.columns, q)
delta_antonov = remove_outliers(delta_antonov, delta_antonov.columns, q)
delta_antonov[1] = np.mean(delta_antonov[[.95, 1.05]].mean())
```

```python
ax = plt.subplot()
ax.plot(delta_antonov.mean(), color=colors[0])
ax.plot(delta_ANN.mean(), color=colors[3], ls='-.')
ax.plot(delta_hagan.mean(), color=colors[1], ls='--')

# plt.ylim(0,.25)
plt.xlim(.4,1.6)
plt.ylim(0,None)

plt.ylabel('Relative error')
plt.xlabel('Strike')
ax.xaxis.set_minor_locator(MultipleLocator(.05))
# ax.yaxis.set_minor_locator(MultipleLocator(.01))

# ax.xaxis.set_major_locator(MultipleLocator(.1))
# ax.yaxis.set_major_formatter(mtick.PercentFormatter())
yvals = ax.get_yticks()
ax.set_yticklabels(["{:,.0%}".format(y) for y in yvals])

labels = ['Antonov', 'Antonov ANN', 'Hagan']
plt.legend(loc=9, bbox_to_anchor=(0.5, -0.14), labels=labels, ncol=3)

plt.grid(alpha=.5)

# plt.savefig(save_path + '/simulation_error_delta.eps', bbox_inches='tight')
```

## Parameter errors
using emperical cdf

```python
def get_stats_df(df, col, q=.995):
    df[df>df.quantile(q)] = np.nan

    stats_df = df \
    .groupby(col) \
    [col] \
    .agg('count') \
    .pipe(pd.DataFrame) \
    .rename(columns = {col: 'frequency'})

    # PDF
    stats_df['pdf'] = stats_df['frequency'] / sum(stats_df['frequency'])

    # CDF
    stats_df['cdf'] = stats_df['pdf'].cumsum()
    stats_df = stats_df.reset_index()
    
    return stats_df
```

```python
def compute_rho_error(df):
    df['rho_error'] = np.abs(df['rho_real'] - df['rho']) / (1 + np.abs(df['rho_real']))
    return df
```

```python
df_ann = ann['df']
df_hagan = hagan['df']
df_antonov = antonov['df']
```

```python
df_ann = compute_rho_error(df_ann)
df_hagan = compute_rho_error(df_hagan)
df_antonov = compute_rho_error(df_antonov)
```

```python
col = 'alpha_error'
ax = plt.subplot()
q = .995

ann_stats = get_stats_df(df_ann, col, q)
hagan_stats = get_stats_df(df_hagan, col, q)
antonov_stats = get_stats_df(df_antonov, col, q)

# plt.plot(ann_stats[col], ann_stats['cdf'])
# plt.plot(hagan_stats[col], hagan_stats['cdf'])

plt.plot(antonov_stats['cdf'], antonov_stats[col])
plt.plot(ann_stats['cdf'], ann_stats[col])
plt.plot(hagan_stats['cdf'], hagan_stats[col])

# plt.ylim(0,1)

# plt.plot(antonov_stats['cdf'], antonov_stats[col])
plt.grid(alpha=.5)
```

```python
col = 'rho_error'
ax = plt.subplot()
q = .95

ann_stats = get_stats_df(df_ann, col, q)
hagan_stats = get_stats_df(df_hagan, col, q)
antonov_stats = get_stats_df(df_antonov, col, q)

# plt.plot(ann_stats[col], ann_stats['cdf'])
# plt.plot(hagan_stats[col], hagan_stats['cdf'])

plt.plot(antonov_stats['cdf'], antonov_stats[col])
plt.plot(ann_stats['cdf'], ann_stats[col])
plt.plot(hagan_stats['cdf'], hagan_stats[col])

plt.ylim(0,4)

# plt.plot(antonov_stats['cdf'], antonov_stats[col])
plt.grid(alpha=.5)
```
